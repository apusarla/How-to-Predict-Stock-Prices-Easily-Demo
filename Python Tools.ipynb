```python
def get_better_colname(col):
    special_symbols = ['&',',','.','(',')']
    
    for s in special_symbols:
        col = col.replace(s, '')
        
    col = col.replace('-','_')
    
    col = col.upper().strip()
    
    col_split = col.split(' ')
    
    col_split_clean = []
    
    for part in col_split:
        if not part.strip() == '':
            col_split_clean.append(part.strip())
            
    col_new = '_'.join(col_split_clean)
    
    return col_new

get_better_colname('Depreciation & Amortization')
    
```

```python
col_map = {}
col_new_list =[]

col_list= df.columns.tolist()

for col in col_list:
    col_new = get_better_colname(col)
    col_map[col] = col_new
    col_new_list.append(col_new)
    
print(col_new_list)

col_map
```

```python
df_profit_2019_q3 = df_profit_2019_q3.copy()
```

```python
df_profit_2019_q3['EPS'] = df_profit_2019_q3.NET_INCOME / df_profit_2019_q3.SHARES_BASIC
df_profit_2019_q3['GROSS_PROFIT_MARGIN'] = df_profit_2019_q3.GROSS_PROFIT / df_profit_2019_q3.REVENUE * 100
df_profit_2019_q3['PROFIT_MARGIN'] = df_profit_2019_q3.NET_INCOME / df_profit_2019_q3.REVENUE * 100

df_profit_2019_q3[['TICKER','EPS','GROSS_PROFIT_MARGIN','PROFIT_MARGIN']]
```

```python
df_profit_2019_q3.EPS.describe()
```

```python
## function for checking if its above or below mean

def classify_eps(eps):
    eps_class = 'Below Average'
    
    if eps > 2.67:
        eps_class = 'Above Average'
        
    
    return eps_class
```

```python
df_profit_2019_q3['EPS_CLASS'] = df_profit_2019_q3.EPS.apply(classify_eps)
```

```python
df_profit_2019_q3[['TICKER','EPS_CLASS']]
```

```python
## Decile rank scores bu using qcut


df_profit_2019_q3['EPS_SCORE'] = pd.qcut(df_profit_2019_q3.EPS, 10, labels=False)
df_profit_2019_q3['NET_INCOME_SCORE'] = pd.qcut(df_profit_2019_q3.NET_INCOME, 10, labels=False)
df_profit_2019_q3['PROFIT_MARGIN_SCORE'] = pd.qcut(df_profit_2019_q3.PROFIT_MARGIN, 10, labels=False)

df_profit_2019_q3[['TICKER','EPS_SCORE','NET_INCOME_SCORE','PROFIT_MARGIN_SCORE']]
```

```python
df_profit_2019_q3['AVG_SCORE'] = (
    df_profit_2019_q3.EPS_SCORE +
    df_profit_2019_q3.NET_INCOME_SCORE +
    df_profit_2019_q3.PROFIT_MARGIN_SCORE
) / 3

df_profit_2019_q3.sort_values('AVG_SCORE', ascending=False).head(10)[['TICKER','AVG_SCORE']]
```

```python
def classify_company(row):
    company_class='Below Average'
    
    if row['EPS_SCORE'] > 4 and row['NET_INCOME_SCORE'] > 4 and row['PROFIT_MARGIN_SCORE'] > 4 :
        company_class='Above Average'
    
    return company_class

df_profit_2019_q3['COMPANY_CLASS'] = df_profit_2019_q3.apply(classify_company, axis=1)

df_profit_2019_q3[['TICKER','COMPANY_CLASS']]
```

```python
## Z score z = (x-mean)/ sd

# The formula implies that Z score can be either positive or negative. Positive Z score means the company is better 
# than the average and negative means the opposite.

eps_mean = df_profit_2019_q3.EPS.mean()
eps_std = df_profit_2019_q3.EPS.std()

df_profit_2019_q3['EPS_Z_SCORE'] = (df_profit_2019_q3.EPS - eps_mean) / eps_std

net_income_mean = df_profit_2019_q3.NET_INCOME.mean()
net_income_std = df_profit_2019_q3.NET_INCOME.std()

df_profit_2019_q3['NET_INCOME_Z_SCORE'] = (df_profit_2019_q3.NET_INCOME - net_income_mean) / net_income_std

profit_margin_mean = df_profit_2019_q3.PROFIT_MARGIN.mean()
profit_margin_std = df_profit_2019_q3.PROFIT_MARGIN.std()

df_profit_2019_q3['PROFIT_MARGIN_Z_SCORE'] = (df_profit_2019_q3.PROFIT_MARGIN - profit_margin_mean) / profit_margin_std

df_profit_2019_q3['AVG_Z_SCORE'] = (
    df_profit_2019_q3.EPS_Z_SCORE +
    df_profit_2019_q3.NET_INCOME_Z_SCORE +
    df_profit_2019_q3.PROFIT_MARGIN_Z_SCORE
) / 3

df_profit_2019_q3.sort_values('AVG_Z_SCORE', ascending=False).head(10)[['TICKER','AVG_Z_SCORE']]
```

```python


df_profit_2019_q3[['TICKER','EPS_Z_SCORE']].head()
```

```python
###One of the main disadvantages of Z score is the potential extreme values. 
###Unlike decile rank (capped between 0 and 9), Z score can have very large numbers on either side. 
###So usually people adjust it with a process called winsorize, which means cap the maximum and minimum values 
###to a given level. We can improve our model as following.

df_profit_2019_q3.EPS_Z_SCORE = df_profit_2019_q3.EPS_Z_SCORE.clip(-2,2)

df_profit_2019_q3.NET_INCOME_Z_SCORE = df_profit_2019_q3.NET_INCOME_Z_SCORE.clip(-2,2)

df_profit_2019_q3.PROFIT_MARGIN_Z_SCORE = df_profit_2019_q3.PROFIT_MARGIN_Z_SCORE.clip(-2,2)

df_profit_2019_q3['AVG_Z_SCORE'] = (
    df_profit_2019_q3.EPS_Z_SCORE +
    df_profit_2019_q3.NET_INCOME_Z_SCORE +
    df_profit_2019_q3.PROFIT_MARGIN_Z_SCORE
) / 3

df_profit_2019_q3.sort_values('AVG_Z_SCORE', ascending=False).head(10)[['TICKER','AVG_Z_SCORE']]


```

```python
cond_1 = ~df_profit_sector.SECTOR.isin(['Basic Materials','Energy','Financial Services'])

df_profit_sector_sub = df_profit_sector[cond_1].copy()

df_profit_sector_sub
```

```python
# now calulate z score

df_profit_sector_mean = df_profit_sector_sub.groupby('SECTOR').mean()
df_profit_sector_std = df_profit_sector_sub.groupby('SECTOR').std()

#df_profit_2019_q3['EPS_Z_SCORE'] = (df_profit_2019_q3.EPS - eps_mean) / eps_std
```

```python
df_profit_sector_mean
```

```python
df_profit_sector_std
```

```python
df_profit_sector_sub = pd.merge(
    left=df_profit_sector_sub,
    right=df_profit_sector_mean,
    how='left',
    on='SECTOR',
    suffixes=['','_MEAN']
)
```

```python
df_profit_sector_sub = pd.merge(
    left=df_profit_sector_sub,
    right=df_profit_sector_std,
    how='left',
    on='SECTOR',
    suffixes=['','_STD']
)
```

```python
df_profit_sector_sub.head()
```

```python
df_profit_sector_sub.columns
```

```python
df_profit_sector_sub['EPS_Z'] = (df_profit_sector_sub.EPS - df_profit_sector_sub.EPS_MEAN) / df_profit_sector_sub.EPS_STD

df_profit_sector_sub['NET_INCOME_Z'] = (
    df_profit_sector_sub.GROSS_PROFIT_MARGIN - df_profit_sector_sub.GROSS_PROFIT_MARGIN_MEAN
) / df_profit_sector_sub.GROSS_PROFIT_MARGIN_STD

df_profit_sector_sub['PROFIT_MARGIN_Z'] = (
    df_profit_sector_sub.PROFIT_MARGIN - df_profit_sector_sub.PROFIT_MARGIN_MEAN
) / df_profit_sector_sub.PROFIT_MARGIN_STD



df_profit_sector_sub['Z'] = (
    df_profit_sector_sub.EPS_Z +
    df_profit_sector_sub.NET_INCOME_Z +
    df_profit_sector_sub.PROFIT_MARGIN_Z
) / 3

```

```python
df_top_10 = df_profit_sector_sub.sort_values('Z', ascending=False).head(10)
df_top_10
```

```python
#
##Explore the top 10 earnings trend
###In most of the cases, the trend is one of the important factors to consider when picking up stocks, therefore we may have heard of Growth Factor a lot. We could have added growth factor into our model, for example, the percentage change of Net Income. But on the other hand, we could visualize it with the Pandas function “pivot_table” too.
###Let’s do it together with an example. Say we want to explore the top 10 stocks’ quarterly earnings trend from 2018 Q1. With the last piece of code, we have got the top 10 list, so we can filter the DataFrame “df” (the original company financial table read from the CSV at the very beginning stage) to those ten stocks only.

cond_top_10 = df.TICKER.isin(df_top_10.TICKER)
cond_year = df.FISCAL_YEAR >= 2018

df_top_10_hist = df[cond_top_10 & cond_year]
df_top_10_hist.head(10)
```

```python
## use PIVOT TABLE

df_top_10_net_income = df_top_10_hist.pivot_table(
    index=['FISCAL_YEAR','FISCAL_PERIOD'],
    columns=['TICKER'],
    values='NET_INCOME'
)
df_top_10_net_income
```
